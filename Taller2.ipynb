Graficar la ecuaci√≥n

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la funci√≥n
def f(x):
    return x**3 - 3*x**2 + x - 1

# Crear un rango de valores de x
x = np.linspace(-2, 4, 400)
y = f(x)

# Dibujar la gr√°fica
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fica de la ecuaci√≥n')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

 Implementar el M√©todo de Newton

```python
from scipy.optimize import newton

# Definir la derivada de la funci√≥n
def df(x):
    return 3*x**2 - 6*x + 1

# Funci√≥n para encontrar ra√≠ces y retornar las estimaciones visitadas
def newton_with_estimates(func, x0, fprime, tol=1.48e-08, maxiter=50):
    estimates = []
    for _ in range(maxiter):
        fx = func(x0)
        dfx = fprime(x0)
        if abs(fx) < tol:  # Si la funci√≥n es suficientemente peque√±a, se detiene
            break
        x0 = x0 - fx / dfx  # Actualizar x0 usando el m√©todo de Newton
        estimates.append(x0)  # Guardar la estimaci√≥n
    return x0, estimates
```

Probar el M√©todo de Newton con Diferentes Estimaciones Iniciales

```python
# Probar con 3 estimaciones iniciales diferentes
initial_guesses = [-1, 0.5, 2.5]
roots = []
all_estimates = []

for guess in initial_guesses:
    root, estimates = newton_with_estimates(f, guess, df)
    roots.append(root)
    all_estimates.append(estimates)

# Mostrar las ra√≠ces encontradas
print(f"Las ra√≠ces encontradas son: {roots}")
print(f"Todas las estimaciones visitadas: {all_estimates}")
```

Graficar la Funci√≥n y las Estimaciones Visitadas

```python
# Dibujar la gr√°fica
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1', color='orange')

# Colores para las estimaciones
colors = ['red', 'green', 'blue']

# Graficar las estimaciones visitadas
for i, estimates in enumerate(all_estimates):
    plt.scatter(estimates, [f(est) for est in estimates], color=colors[i], label=f'Estimaciones desde {initial_guesses[i]}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fica de la ecuaci√≥n y = x^3 - 3x^2 + x - 1')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
