Graficar la ecuación

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función
def f(x):
    return x**3 - 3*x**2 + x - 1

# Crear un rango de valores de x
x = np.linspace(-2, 4, 400)
y = f(x)

# Dibujar la gráfica
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la ecuación')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

 Implementar el Método de Newton

```python
from scipy.optimize import newton

# Definir la derivada de la función
def df(x):
    return 3*x**2 - 6*x + 1

# Función para encontrar raíces y retornar las estimaciones visitadas
def newton_with_estimates(func, x0, fprime, tol=1.48e-08, maxiter=50):
    estimates = []
    for _ in range(maxiter):
        fx = func(x0)
        dfx = fprime(x0)
        if abs(fx) < tol:  # Si la función es suficientemente pequeña, se detiene
            break
        x0 = x0 - fx / dfx  # Actualizar x0 usando el método de Newton
        estimates.append(x0)  # Guardar la estimación
    return x0, estimates
```

Probar el Método de Newton con Diferentes Estimaciones Iniciales

```python
# Probar con 3 estimaciones iniciales diferentes
initial_guesses = [-1, 0.5, 2.5]
roots = []
all_estimates = []

for guess in initial_guesses:
    root, estimates = newton_with_estimates(f, guess, df)
    roots.append(root)
    all_estimates.append(estimates)

# Mostrar las raíces encontradas
print(f"Las raíces encontradas son: {roots}")
print(f"Todas las estimaciones visitadas: {all_estimates}")
```

Graficar la Función y las Estimaciones Visitadas

```python
# Dibujar la gráfica
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1', color='orange')

# Colores para las estimaciones
colors = ['red', 'green', 'blue']

# Graficar las estimaciones visitadas
for i, estimates in enumerate(all_estimates):
    plt.scatter(estimates, [f(est) for est in estimates], color=colors[i], label=f'Estimaciones desde {initial_guesses[i]}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la ecuación y = x^3 - 3x^2 + x - 1')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
